```python
import os
import pandas as pd
import PyPDF2
import re

# Define the folder containing your CSV files
folder_path = './'

# List all CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.pdf')]

text = []
# Iterate through the CSV files and run your model on each one
for csv_file in csv_files:
    file_path = os.path.join(folder_path, csv_file)
    # Load the CSV file into a DataFrame
    pdf = PyPDF2.PdfFileReader(file_path)

    # Initialize an empty string to store the text
    pdf_text = ""

    # Loop through each page and extract text
    for page_num in range(pdf.numPages):
        page = pdf.getPage(page_num)
        page_text = page.extractText()
        pdf_text += page_text

    # Close the PDF file
    # pdf.close()

    # Remove special characters and extra spaces
    cleaned_text = re.sub(r'[^\w\s/:-]', '', pdf_text)

    # Replace multiple spaces and line breaks with a single space
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)

    # Strip leading and trailing spaces
    cleaned_text = cleaned_text.strip()
    text.append(cleaned_text)
```

```python
df = pd.DataFrame(text)
df
```

```python
df.to_csv('resume_text.csv')
```
